```python

def BinaryToDecimal(binary):
		
	binary1 = binary
	decimal, i, n = 0, 0, 0
	while(binary != 0):
		dec = binary % 10
		decimal = decimal + dec * pow(2, i)
		binary = binary//10
		i += 1
	return (decimal)

```

```python
name = input("Enter any text: ")
key=int(input("enter int key value"))
Bin_Name = ''.join(format(ord(i), '08b') for i in name)
```

```python
x=0
y=0
```

```python
#encryption of colored image
for j in Bin_Name:

  temp= bin(img_encrypted[x][y])
  # temp= format(img_c[x][y][0], "b")
  l=len(temp)
  # print(temp,end=" ")
  
  tmp=temp[:l-1]+j;
  # print(tmp)
  img_encrypted[x][y]=int(tmp, 2)
  
  y+=key
  if(y>222):
    x+=y//223
    x=x%222
    y=y%223
```

```python
x_c=0;
y_c=0
```

```python
#decryption of colored image
str_name=""

for i in range(len(Bin_Name)):
  temp= bin(img_encrypted[x_c][y_c])
  l=len(temp)
  # print(img_c[x][y][0])
  # temp= format(img_c[x][y][0], "b")
  str_name=str_name+temp[l-1]
  y_c+=key;
  if(y_c>222):
    x_c+=y_c//223;
    x_c=x_c%222
    y_c=y_c%223
# print(str_name)
hidden_name =''
for i in range(0, len(str_name), 8):
	
	temp_data = int(str_name[i:i + 8])
	
	decimal_data = BinaryToDecimal(temp_data)
	hidden_name = hidden_name + chr(decimal_data)

	

print("hidden name is  :",hidden_name)
```
